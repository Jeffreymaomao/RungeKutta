# Runge-Kutta methods in python
## The ODE problem
 The initial value problem is
     $$\frac{d\vec{y}}{dx}=\vec{f}\left(x,\vec{y}\right),\quad \vec{y}\left(x_0\right)=y_0,$$
 here $\vec{y}$ is an unknown function (scalar or vector), which we would like to approximate.
## The Explicit Runge–Kutta methods
 The Explicit Runge–Kutta methods for $s$-stage take the form
     $$\vec{y}_{n+1}=\vec{y}_n+h\sum_{i=1}^{s}b_i\vec{f}_i,$$
 where
    $$\begin{aligned}
    \vec{f}_1&=\vec{f}\left(x_n\ ,\ \vec{y}_n\right)
    \\
    \vec{f}_2&=\vec{f}\left(x_n+c_2h\ ,\ \vec{y}_n+ha_{21}\vec{f}_1\right)
    \\
    \vec{f}_3&=\vec{f}\left(x_n+c_3h\ ,\ \vec{y}_n+h\left(a_{31}\vec{f}_1+a_{32}\vec{f}_2\right)\right)
    \\
    &\vdots
    \\
    \vec{f}_s&=\vec{f}\left(x_n+c_sh\ ,\ \vec{y}_n+h\left(a_{s1}\vec{f}_1+a_{s2}\vec{f}_2+\cdots a_{s,s-1}\vec{f}_{s-1}\right)\right)
    \end{aligned}$$
## python code

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
```

### 1 stage Runge-Kutta method
The method is also called Euler method
$$\vec{y}_{n+1}=\vec{y}_n+h\vec{f}\left(x_n,\vec{y}_n\right)$$

```python
def RK1(f,y_0,x,h):
    y = np.zeros((len(x),len(y_0)))
    y[0] = y_0
    for i in range(0,len(x)-1):
        y[i+1] = y[i] + h * f(x[i],y[i])
    return y
```

### 2 stage Runge-Kutta method
The method is given by
$$\vec{y}_{n+1}=\vec{y}_n+h\left(b_1\vec{f}_1+b_2\vec{f}_2\right)$$
where
$$\begin{aligned}
\vec{f}_1 & = \vec{f}\left(x_n,\vec{y}_n\right)\\
\vec{f}_2 & = \vec{f}\left(x_n+c_2h,\vec{y}_n+ha_{21}\vec{f}_1\right)
\end{aligned}$$

```python
def RK2(f,y_0,x,h):
    b1, b2 = 1/2, 1/2
    c1, c2 = 0, 1
    a21 = 1
    y = np.zeros((len(x),len(y_0)))
    y[0] = y_0
    for i in range(0,len(x)-1):
        f1 = f(x[i] + c1*h, y[i])
        f2 = f(x[i] + c2*h, y[i] + h*a21*f1)
        y[i+1] = y[i] + h * (b1*f1 + b2*f2)
    return y
```

### 3 stage Runge-Kutta method
The method is given by
$$\vec{y}_{n+1}=\vec{y}_n+h\left(b_1\vec{f}_1+b_2\vec{f}_2+b_3\vec{f}_3\right)$$
where
$$\begin{aligned}
\vec{f}_1 & = \vec{f}\left(x_n,\vec{y}_n\right)\\
\vec{f}_2 & = \vec{f}\left(x_n+c_2h,\vec{y}_n+ha_{21}\vec{f}_1\right)\\
\vec{f}_3 & = \vec{f}\left(x_n+c_3h,\vec{y}_n+h\left(a_{31}\vec{f}_1+a_{32}\vec{f}_2\right)\right)
\end{aligned}$$

```python
def RK3(f,y_0,x,h):
    b1, b2, b3 = 1/6, 2/3, 1/6
    c1, c2, c3 = 0, 1/2, 1
    a21 = 1/2
    a31, a32 = -1, 2
    y = np.zeros((len(x),len(y_0)))
    y[0] = y_0
    for i in range(0,len(x)-1):
        f1 = f(x[i] + c1*h, y[i])
        f2 = f(x[i] + c2*h, y[i] + h*a21*f1)
        f3 = f(x[i] + c3*h, y[i] + h*a31*f1 + h*a32*f2)
        y[i+1] = y[i] + h * (b1*f1 + b2*f2 + b3*f3)
    return y
```

### 4 stage Runge-Kutta method
The method is given by
$$\vec{y}_{n+1}=\vec{y}_n+h\left(b_1\vec{f}_1+b_2\vec{f}_2+b_3\vec{f}_3+b_4\vec{f}_4\right)$$
where
$$\begin{aligned}
\vec{f}_1 & = \vec{f}\left(x_n,\vec{y}_n\right)\\
\vec{f}_2 & = \vec{f}\left(x_n+c_2h,\vec{y}_n+ha_{21}\vec{f}_1\right)\\
\vec{f}_3 & = \vec{f}\left(x_n+c_3h,\vec{y}_n+h\left(a_{31}\vec{f}_1+a_{32}\vec{f}_2\right)\right)\\
\vec{f}_4 & = \vec{f}\left(x_n+c_4h,\vec{y}_n+h\left(a_{41}\vec{f}_1+a_{42}\vec{f}_2+a_{43}\vec{f}_3\right)\right)
\end{aligned}$$

```python
def RK4(f,y_0,x,h):
    b1, b2, b3, b4 = 1/6, 1/3, 1/3, 1/6
    c1, c2, c3, c4 = 0, 1/2, 1/2, 1
    a21           = 1/2
    a31, a32      = 0, 1/2
    a41, a42, a43 = 0, 0, 1
    y = np.zeros((len(x),len(y_0)))
    y[0] = y_0
    for i in range(0,len(x)-1):
        f1 = f(x[i] + c1*h, y[i])
        f2 = f(x[i] + c2*h, y[i] + h*a21*f1)
        f3 = f(x[i] + c3*h, y[i] + h*a31*f1 + h*a32*f2)
        f4 = f(x[i] + c4*h, y[i] + h*a41*f1 + h*a42*f2 + h*a43*f3)
        y[i+1] = y[i] + h * (b1*f1 + b2*f2 + b3*f3 + b4*f4)
    return y
```

### 5 stage Runge-Kutta method
The method is given by
$$\vec{y}_{n+1}=\vec{y}_n+h\left(b_1\vec{f}_1+b_2\vec{f}_2+b_3\vec{f}_3+b_4\vec{f}_4+b_5\vec{f}_5\right)$$
where
$$\begin{aligned}
\vec{f}_1 & = \vec{f}\left(x_n,\vec{y}_n\right)\\
\vec{f}_2 & = \vec{f}\left(x_n+c_2h,\vec{y}_n+ha_{21}\vec{f}_1\right)\\
\vec{f}_3 & = \vec{f}\left(x_n+c_3h,\vec{y}_n+h\left(a_{31}\vec{f}_1+a_{32}\vec{f}_2\right)\right)\\
\vec{f}_4 & = \vec{f}\left(x_n+c_4h,\vec{y}_n+h\left(a_{41}\vec{f}_1+a_{42}\vec{f}_2+a_{43}\vec{f}_3\right)\right)\\
\vec{f}_5 & = \vec{f}\left(x_n+c_5h,\vec{y}_n+h\left(a_{51}\vec{f}_1+a_{52}\vec{f}_2+a_{53}\vec{f}_3+a_{54}\vec{f}_4\right)\right)
\end{aligned}$$

```python
def RK5(f,y_0,x,h):
    b1 ,b2 ,b3 ,b4, b5  = 1/6, 0, 0, 2/3, 1/6
    c1, c2, c3, c4, c5  = 0, 1/3, 1/3, 1/2, 1
    a21                 = 1/3  
    a31, a32            = 1/6,  1/6
    a41, a42, a43       = 1/8,  0,   3/8
    a51, a52, a53, a54  = 1/2,  0,   -3/2, 2
    y = np.zeros((len(x),len(y_0)))
    y[0] = y_0
    for i in range(0,len(x)-1):
        f1 = f(x[i] + c1*h, y[i])
        f2 = f(x[i] + c2*h, y[i] + h*a21*f1)
        f3 = f(x[i] + c3*h, y[i] + h*a31*f1 + h*a32*f2)
        f4 = f(x[i] + c4*h, y[i] + h*a41*f1 + h*a42*f2 + h*a43*f3)
        f5 = f(x[i] + c5*h, y[i] + h*a51*f1 + h*a52*f2 + h*a53*f3 + h*a54*f4)
        y[i+1] = y[i] + h * (b1*f1 + b2*f2 + b3*f3 + b4*f4 + b5*f5)
    return y
```

## Example

### 2nd order ODE
The ODE is given by
$$\frac{d^2y}{dx^2} = \vec{f}\left(x,\vec{y},\vec{y}'\right)$$

then we define the vector form ODE
$$\frac{d\vec{Y}}{dx} = \frac{d}{dx}\begin{pmatrix}y\\y'\end{pmatrix} 
    =  \begin{pmatrix}y'\\y''\end{pmatrix} 
    =  \begin{pmatrix}y'\\\vec{f}\left(x,\vec{y},\vec{y}'\right)\end{pmatrix}
    \equiv\vec{F}\left(x,\vec{Y}\right).$$
    
Taking
$$y''=-2y'-2y, \quad y\left(0\right)=2, y'\left(0\right)=1$$

as a example, first we define $\vec{F}$ function and initial vector $\vec{Y}_0 = \begin{pmatrix}2\\1\end{pmatrix}$

```python
def F(x,Y):
    y0, y1 = Y[0], Y[1]
    F0 = y1
    F1 = -2*y1 -2*y0
    return np.array((F0,F1))
Y0 = np.array((2,1))
```

For $x\in[0,1)$ we define $x=[0,h,\ldots,10]$ with the step $h$. Then solving the initial value with big step $h=0.6$

```python
h = 0.6
x = np.arange(0,10,h)
y1 = RK1(F,Y0,x,h)
y2 = RK2(F,Y0,x,h)
y3 = RK3(F,Y0,x,h)
y4 = RK4(F,Y0,x,h)
y5 = RK5(F,Y0,x,h)
```

The explicit solution of this ODE is 
$$y\left(x\right)=e^{-x}\left(3\sin(x)+2\cos(x)\right)$$
and
$$y'\left(x\right)=e^{-x}\left(\cos(x)-5\sin(x)\right)$$

```python
def y0_sol(x):
    return np.e**(-x)*(3*np.sin(x) + 2*np.cos(x))
def y1_sol(x):
    return np.e**(-x)*(np.cos(x) - 5*np.sin(x))
```

then plot the solution $\left(x,y\right)$ and $\left(x,y'\right)$

```python
fig, ax = plt.subplots(2, 3,figsize=(15, 10), dpi = 200)

ax[0, 0].plot(x, y0_sol(x), 'red', label = r'$y$')
ax[0, 0].plot(x, y1_sol(x), 'blue', label = r'${dy}/{dx}$')
ax[0, 0].legend()

ax[0, 1].plot(x, y1[:,0], 'red', label = r'$y$')
ax[0, 1].plot(x, y1[:,1], 'blue', label = r'${dy}/{dx}$')
ax[0, 1].legend()

ax[0, 2].plot(x, y2[:,0], 'red', label = r'$y$')
ax[0, 2].plot(x, y2[:,1], 'blue', label = r'${dy}/{dx}$')
ax[0, 2].legend()

ax[1, 0].plot(x, y3[:,0], 'red', label = r'$y$')
ax[1, 0].plot(x, y3[:,1], 'blue', label = r'${dy}/{dx}$')
ax[1, 0].legend()

ax[1, 1].plot(x, y4[:,0], 'red', label = r'$y$')
ax[1, 1].plot(x, y4[:,1], 'blue', label = r'${dy}/{dx}$')
ax[1, 1].legend()

ax[1, 2].plot(x, y4[:,0], 'red', label = r'$y$')
ax[1, 2].plot(x, y4[:,1], 'blue', label = r'${dy}/{dx}$')
ax[1, 2].legend()

ax[0, 0].set_title("explicit solution")
ax[0, 1].set_title("Runge-Kutta 1")
ax[0, 2].set_title("Runge-Kutta 2")
ax[1, 0].set_title("Runge-Kutta 3")
ax[1, 1].set_title("Runge-Kutta 4")
ax[1, 2].set_title("Runge-Kutta 5")

fig.tight_layout()
plt.show()
```
